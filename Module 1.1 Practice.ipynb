# Latihan 1: Mean, Median, Modus
1. Diketahui tinggi badan siswa kelas A adalah 100,5 ; 99,7 ; 101 ; 101,3 ; 99,9 ; 98,9 ; 165 ; 102,2 ; 98,7 ; 102,9 ; dan 100. Menggunakan Mean, Median, dan Modus, coba gambarkan pusat distribusi data berikut. Manakah yang paling bagus?

2. Berikut adalah nilai UAS pada mata pelajaran matematika di kelas A: 95, 35, 50, 65, 70, 85, 25, 100, 80, 90, 45, 40, 35, 20, 75, 50, 40, 60, 80, 75. Menggunakan Mean, Median, dan Modus, coba gambarkan pusat distribusi data berikut. Manakah yang paling bagus?

# Nomor 1
# Menghitung Mean:

```python
import statistics

data = [100.5, 99.7, 101, 101.3, 99.9, 98.9, 165, 102.2, 98.7, 102.9, 100]
print(f"Data: {data}")

jumlah = sum(data)
print(f"Jumlah : {jumlah: .2f}")

banyak_data = len(data)
print(f"Banyak Data: {banyak_data}")

mean = jumlah / banyak_data
print(f"Mean: {mean: .2f}")
```

```python
import matplotlib.pyplot as plt

plt.hist(data, bins=30, edgecolor='black', alpha=0.5, color='blue')

plt.axvline(x=mean, color='red', linestyle='dashed', linewidth=1, label=f'Mean = {mean:.2f}')

plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Mean')

plt.legend()

plt.show()
```

# Menghitung Median:

```python
import statistics

data = [100.5 , 99.7 , 101 , 101.3 , 99.9 , 98.9 , 165 , 102.2 , 98.7 , 102.9 , 100]
print(f"Data: {data}")

sorted_data = sorted(data)
print(f"Data berurut: {sorted_data}")

n = len(sorted_data)
print(f"Jumlah total data: {n}")

Tengah = n // 2

# Jika jumlah data genap
if n % 2 == 0:
    print(f"Jumlah total data: Genap")
    median1 = sorted_data[Tengah]
    median2 = sorted_data[Tengah - 1]
    print(f"Nilai tengah: {median1} {median2}")
    median = (median1 + median2) / 2

# Jika jumlah data ganjil
else:
    print(f"Jumlah total data: Ganjil")
    median = sorted_data[Tengah]
    print(f"Nilai tengah: {median}")

print(f"Median: {median}")
```

# Menghitung Modus:

```python
from collections import Counter

data = [100.5, 99.7, 101, 101.3, 99.9, 98.9, 165, 102.2, 98.7, 102.9, 100]
print(f"Data: {data}")

counter = Counter(data)
mode_data = dict(counter)
max_counter = max(list(counter.values())) 
print(f"Jumlah paling banyak: {max_counter}")

mode_value = []
for k, v in mode_data.items():
    if v == max(list(counter.values())):
        mode_value.append(k)
    
print(f"Modus: {mode_value}")
```

```python
import matplotlib.pyplot as plt

plt.hist(data, bins=30, edgecolor='black', alpha=0.5, color='blue')

plt.axvline(x=median, color='red', linestyle='dashed', linewidth=1, label=f'Median = {median:.2f}')

plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Median')

plt.legend()

plt.show()
```

# Kesimpulan Nomor 1:
Berdasarkan analisis grafik dan data, terdapat perbedaan pendapat tentang pemusatan data yang paling tepat. Median dianggap lebih baik karena mendekati frekuensi tertinggi data, sementara mean dianggap mewakili seluruh data meskipun ada nilai yang jauh dari kumpulan data lainnya. Oleh karena itu, pemilihan antara median atau mean sebagai pusat distribusi terbaik bergantung pada konteks dan karakteristik data tersebut.

# Menghitung Modus:

```python
from collections import Counter

data = [95, 35, 50, 65, 70, 85, 25, 100, 80, 90, 45, 40, 35, 20, 75, 50, 40, 60, 80, 75]
print(f"Data: {data}")

counter = Counter(data)
mode_data = dict(counter)
max_counter = max(list(counter.values())) 
print(f"Jumlah paling banyak: {max_counter}")

mode_value = []
for k, v in mode_data.items():
    if v == max(list(counter.values())):
        mode_value.append(k)
    
print(f"Modus: {mode_value}")
```

```python
import matplotlib.pyplot as plt

# Create a histogram
plt.hist(data, bins=30, edgecolor='black', alpha=0.5, color='green')

# Add a vertical line at the mean
plt.axvline(x=mode_value[0], color='red', linestyle='dashed', linewidth=1, label=f'Modus 1 = {mode_value[0]:.2f}')
plt.axvline(x=mode_value[1], color='blue', linestyle='dashed', linewidth=1, label=f'Modus 2 = {mode_value[1]:.2f}')
plt.axvline(x=mode_value[2], color='yellow', linestyle='dashed', linewidth=1, label=f'Modus 3 = {mode_value[2]:.2f}')
plt.axvline(x=mode_value[3], color='cyan', linestyle='dashed', linewidth=1, label=f'Modus 4 = {mode_value[3]:.2f}')
plt.axvline(x=mode_value[4], color='pink', linestyle='dashed', linewidth=1, label=f'Modus 5 = {mode_value[4]:.2f}')

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Modes')

# Add legend
plt.legend()

# Show the plot
plt.show()
```

Oleh karena itu, pemilihan antara modus atau median sebagai pusat distribusi terbaik bergantung pada karakteristik dan konteks data tersebut.
