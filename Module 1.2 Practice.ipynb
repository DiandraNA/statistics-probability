# Latihan 2: Range, Standar Deviasi
1. Dalam suatu liga sepak bola, dipilih secara acak 10 pemain dari klub yang berbeda dan diketahui jumlah gol yang mereka raih dalam musim ini adalah: 19, 10, 10, 12, 13, 15, 14, 13, 9, 17. Lakukan analisis terhadap sebaran data dari dataset yang diberikan. Manakah ukuran sebaran yang paling mampu menggambarkan sebaran data dengan baik?

2. Dalam tim olahraga, tinggi pemain adalah: 180 cm, 185 cm, 175 cm, 190 cm, 170 cm, 195 cm, 165 cm. Bagaimana sebaran data dari tinggi pemain ini?

# Nomor 1
# Mencari Range:

```python
def range(data):
    nilai_tertinggi = max(data)
    print(f"Nilai tertinggi: {nilai_tertinggi}")

    nilai_terendah = min(data)
    print(f"Nilai terendah: {nilai_terendah}")

    data_range = max(data) - min(data)
 
    print(f"Range: {nilai_tertinggi} - {nilai_terendah} = {data_range}")  
    
    return data_range


data = [19, 10, 10, 12, 13, 15, 14, 13, 9, 17]
print(f"Data: {data}")

data_range = range(data)
```

```python
import matplotlib.pyplot as plt

def range(data):
    nilai_tertinggi = max(data)
    nilai_terendah = min(data)
    data_range = max(data) - min(data)
    return data_range

plt.hist(data, bins=30, edgecolor='black', alpha=0.5, color='yellow')

y = 1.5 # The y-coordinate where the line will be drawn
x1 = 4/70*1  # The x-coordinate of the start point
x2 = 66/70*1  # The x-coordinate of the end point


plt.axhline(y=y, xmin=x1, xmax=x2, color='red', linestyle='dashed', linewidth=1, label=f'Range = {range(data)}')

plt.axvline(x=min(data), color='blue', linestyle='dashed', linewidth=1, label=f'Min Data: {min(data):.2f}')
plt.axvline(x=max(data), color='blue', linestyle='dashed', linewidth=1, label=f'Max Data: {max(data):.2f}')


plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Range')


plt.legend()


plt.show()
```

# Mencari Standar Deviasi:

```python
def calculate_mean(data):
    return sum(data) / len(data)

def calculate_variance(data):
    # Rata-rata 
    mean = calculate_mean(data)
    print(f"Rata-rata: {mean}")  
    
    squared_diff = []
    for x in data:
        squared_diff_per_x = (x - mean)**2
        squared_diff.append( squared_diff_per_x ) 
    
    variance = sum(squared_diff) / len(data)
     
    print(f"Variansi: {sum(squared_diff)} / {len(data)} = {variance}")    
    
    return variance

def calculate_standard_deviation(data):
    variance = calculate_variance(data)
    standard_deviation = variance**0.5  
    print(f"Standard Deviation: {variance}**0.5 = {standard_deviation: .2f}")
    
    return standard_deviation

data = [19, 10, 10, 12, 13, 15, 14, 13, 9, 17]
print(f"Data: {data}")

standard_deviation1 = calculate_standard_deviation(data)
mean = calculate_mean(data)
```

```python
import matplotlib.pyplot as plt
import numpy as np


plt.hist(data, bins=30, edgecolor='black', alpha=0.5, color='yellow')


plt.axvline(x=mean, color='red', linestyle='dashed', linewidth=1, label=f'Mean = {mean}')
plt.axvline(x=mean - standard_deviation, color='blue', linestyle='dashed', linewidth=1, label=f'{mean-standard_deviation:.2f}')
plt.axvline(x=mean + standard_deviation, color='blue', linestyle='dashed', linewidth=1, label=f'{mean+standard_deviation:.2f}')


plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title(f'Std Deviation = {standard_deviation:.2f}')


plt.legend()


plt.show()
```

# Kesimpulan Nomor 1:
Berdasarkan analisis grafik dan perhitungan, standar deviasi memberikan gambaran sebaran data yang lebih akurat dibandingkan range, yang hanya menampilkan selisih nilai terbesar dan terkecil. Meskipun data tersebar di antara 9-19 dengan selisih nilai terbesar dan terkecil sebesar 10, standar deviasi 3.02 menunjukkan bahwa data tidak berjarak terlalu jauh satu sama lain. Oleh karena itu, standar deviasi dapat menjadi metode yang lebih tepat untuk menggambarkan sebaran data ini.
